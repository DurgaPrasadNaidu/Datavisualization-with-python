```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib as mlp
import matplotlib.pyplot as plt
%matplotlib  inline
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.xlabel('a',color='blue')
plt.ylabel('b',color='blue')
plt.title('2d-diagram',color='blue')
plt.plot(a,b,'r*',linestyle='dashed',linewidth=2,markersize=12)
```

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175] 
plt.xlabel('days',color='blue')
plt.ylabel('sales of two companies',color='blue')
plt.title('2d-diagram',color='blue')
plt.plot(days,sales_1)
plt.plot(days,sales_2)
```
